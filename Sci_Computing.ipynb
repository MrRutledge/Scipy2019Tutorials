## Slicing and Indexing 

```python
# new a array for SLICING

a = np.array([10,11,12,13,14])
print(a)

print('a[1:3] :', a[1:3])
print('a[1:-2] :', a[1:-2])
print('a[-4:3] :', a[-4:3])
```

#### Omitting  indices

omitted boundieries are assumed to be the beginning or the end of the list 

```python
print(a)
# first element
print('a[:3]-->', a[:3])
#last element 
print('a[-2:] -->', a[-2:])
#every othe element
print('a[::2]-->',a[::2])
```

```python
# slicicng colums
a2 = np.arange(56).reshape(8,7)
a2
```

```python
print(a2)

print('a2[0, 3:5]-->', a2[0,3:5])

print('a2[4:, 4:]-->', a2[4:, 4:])
print('a2[:, 2]-->', a2[:,2])
print('a2[2::2, ::2]-->', a2[2::2, ::2])
```
